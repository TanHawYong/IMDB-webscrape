```python
import numpy as np
import pandas as pd
df = pd.read_csv('movie-dataset.csv')
```

```python
df.head(50)
```

```python
df= df.replace('-', np.nan)
df= df.replace('#[a-z]*', np.nan)
df['Year of release']=df['Year of release'].str.replace("II","")
df['Year of release']=df['Year of release'].str.replace("I","")
df['Genres']=df['Genres'].str.replace(",","/")
df['Production Company']=df['Production Company'].str[1:]
df['Country']=df['Country'].str[1:]
```

```python
df = df[df["Gross Collection"].str.contains("#")==False]
```

```python
df=df.dropna(how='any')
```

```python
df=df.reset_index(drop=True)
```

```python
df.to_csv('movie-dataset-cleaned.csv', index=False )
```
